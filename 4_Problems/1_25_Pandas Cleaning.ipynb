## ðŸŸ© Extract Year from Date (1.25.3)

```python
Problem Statement:
Create a new column job_posted_year that extracts the year from the job_posted_date column and display the first 5 rows.
```

```python
df['job_posted_year']=pd.to_datetime(df['job_posted_date']).dt.year
df['job_posted_year'].head(5)
```

## ðŸŸ¨ Drop Missing Salaries (1.25.4)

```python
Problem Statement:
Drop rows where the salary_year_avg column has missing values and display the first 5 rows of the cleaned DataFrame.
```

```python
df.dropna(subset='salary_year_avg',inplace=True)
df.head(5)
```
