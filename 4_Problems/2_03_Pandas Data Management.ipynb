```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## üü© Drop Missing Values (2.3.1)

```python
# Problem Statement:
# Create a copy of the DataFrame and drop rows where the salary_year_avg column has missing values. Display the first 5 rows of the cleaned DataFrame.
```

```python
df_copy=df.copy()
df_copy.salary_year_avg.dropna().head(5)
df_copy.dropna(subset='salary_year_avg').head(5)
```

## üü® Fill Missing Salaries (2.3.3)

```python
# Problem Statement:
# Create a copy of the original DataFrame and fill in missing values in the salary_year_avg column with the median salary. 
# Display the first 5 rows of the altered DataFrame.
```

```python
df_copy=df.copy()
median_salary=df.salary_year_avg.median()
df_copy.salary_year_avg.fillna(median_salary,inplace=True)
df_copy.head()
```

```python

df_copy = df.copy()
median_salary = df_copy['salary_year_avg'].median()
df_copy['salary_year_avg'] = df_copy['salary_year_avg'].fillna(median_salary)
df_copy.head(5)
```
