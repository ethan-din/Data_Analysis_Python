## üü© Sort Index (2.5.3)

```python
# Problem Statement:
# Create a pivot table to calculate the median salary_year_avg for each company_name and sort the index in descending order. 
# Display the first 5 rows.

# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.pivot_table(values='salary_year_avg',index='company_name',aggfunc='median').sort_index(ascending=False).head()
```

```python
df_pivot_table=df.pivot_table(values='salary_year_avg',index='company_name',aggfunc='median').sort_index(inplace=True,ascending=False)
df_pivot_table.head()

# ËøôÈáåÂ¶ÇÊûú‰ΩøÁî®inplace=trueÔºå‰ºöËøîÂõû‰∏Ä‰∏™Á©∫ÂÄºÔºåÊâÄ‰ª•ÂØºËá¥‰∫ÜÊä•Èîô„ÄÇ‰øÆÊîπÊñπÊ≥ïÔºå‰∏ç‰ΩøÁî®inplaceÊàñËÄÖÈáçÊñ∞ËµãÂÄº
```
