```python
# !pip install -U datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Job Posting Location Bar Plot (1.31.1)

```python
# Problem Statement:
# Create a bar plot with the Pandas .plot() method showing the top 5 job posting platforms (e.g., LinkedIn, Upwork, etc.) 
# with the highest number of job postings. Use the job_via column to count the number of postings for each platform.



axis=df.groupby('job_via').size().sort_values(ascending=False).head()
axis.plot(kind='bar',color='red')
plt.xticks(rotation=45,ha='right')
plt.xlabel('')
plt.show()
```

## ðŸŸ© Top Job Locations (1.31.2)

```python
# Problem Statement:
# Create a horizontal bar plot showing the top 5 job locations with the highest number of job postings:

# Use the job_location column to count the number of postings for each location.
# Add in labels for the x-axis (label as 'Count of Jobs'), y-axis (label as 'Job Location'), and a title (set as 'Number of Job Postings by Job Location') for the chart.



axis=df.groupby('job_location').size().sort_values(ascending=False).head()
axis.plot(kind='barh',color='red')
# plt.xticks(rotation=45,ha='right')
plt.xlabel('Count of Jobs')
plt.ylabel('Job Location')
plt.title('Number of Job Postings by Job Location')
plt.show()


location_counts = df['job_location'].value_counts().head(5)
location_counts.plot(kind='barh')
plt.xlabel('Count of Jobs')
plt.ylabel('Job Location')
plt.title('Number of Job Postings by Job Location')
plt.show()
```

## ðŸŸ¨ Job Postings Trend Line Plot (1.31.3)

```python
# Problem Statement:
# Create a line plot showing the trend of job postings over time using the Pandas method to plot:

# Create a DataFrame called monthly_job_counts and use the job_posted_date column to count the number of job postings for each month (in the format 'Jan').
# Add labels for the x-axis (label as 'Month'), y-axis (label as 'Number of Job Postings'), and a title (set as 'Trend of Job Postings Over Time') for the chart.

df['monthly_job_counts']=df.job_posted_date.dt.to_period('M')
axis=df.groupby('monthly_job_counts').size()
axis.plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.title('Trend of Job Postings Over Time')
plt.show()
```

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['job_posted_month'] = df['job_posted_date'].dt.to_period('M')
monthly_job_counts = df['job_posted_month'].value_counts().sort_index()
monthly_job_counts.plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.title('Trend of Job Postings Over Time')
plt.show()
```
